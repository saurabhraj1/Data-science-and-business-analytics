## Statewise profit and sells: 

```python
pls=df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
pls[:].plot.bar(color=['red','black'],figsize=(20,10))
plt.title('Profit/Loss and Sales across the States')
plt.xlabel('States')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

### Graph Display that California and New York have the highest profits while Texas and Ohio have the least profits.

## PROFIT VS DISCOUNT:

```python
plt.style.use('seaborn')
df.plot(kind="scatter",figsize=(12,6),x="Sales",y="Profit",c="Discount",s=20,fontsize=12,colormap='plasma')
plt.ylabel('Profits')
plt.xlabel('Sales')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

### From graph it is clear that for discount between 0 to 0.5 , discount is negatively related to profits whereas for discounts between 0.5 and 0.6 it has positive relation.

## Regionwise profit and sells: 

```python
pls=df.groupby('Region')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
pls[:].plot.bar(color=['red','black'],figsize=(20,10))
plt.title('Profit/Loss and Sales across the Region')
plt.xlabel('Region')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

### From graph it is clear that western region has highest sells and profits.

## PROFIT VS QUANTITY:

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

## From graph it is clear that as quantity increases profit also increases.

## Profit and sells across different segments: 

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['red','blue'],figsize=(8,5))
plt.ylabel('Profit/loss and Sales')
plt.show()
```

## From graph it is clear that consumer has the highest profit and sales .

## Segment wise sales in each region:

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region',fontsize=15)
plt.show()
```

## Category wise profit and sells: 

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['red','green'],alpha=0.9,figsize=(10,6))
plt.ylabel('Profit/loss and sales')
plt.show()
```

## From graph it is clear that Technology has highest profit and sells.

# Subcategory wise profit and sales: 

```python
ps=df.groupby('Sub-Category')[['Sales','Profit']].mean().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['blue','red'],figsize=(16,10))
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

## From graph it is clear that Copier has highest sale and profit whereas Tables have least profit and Fasteners have least sales.
